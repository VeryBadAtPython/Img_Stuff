```julia
using Images, FileIO

img2 = load("water.jpg")
```

```julia
function thresholdPixel(pixel)
    if Gray(float(pixel)) < 0.1
        return Gray{Bool}(1)
    else
        return Gray{Bool}(0)
    end
end


thresholdPixel.(img2)
```

```julia
function seeRed(x)
    return x.g 
end

Gray.(seeRed.(img2))
```
