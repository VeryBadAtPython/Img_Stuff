```julia
using Images, FileIO

img2 = load("water.jpg")
```

```julia
function thresholdPixel(pixel)
    if Gray(float(pixel)) < 0.4
        return 1
    else
        return 0
    end
end


onlyDark = thresholdPixel.(img2)
Gray.(onlyDark)
```

```julia
function validate(x)
    (r,c) = size(x)
    y = copy(x)
    for i  1:1:r
        for j  1:1:c
            if onlyDark[i,j] == 1
                y[i,j] = RGB{N0f8}(0,0,0)
            end
        end
    end
    return y
end

validate(img2)
```
